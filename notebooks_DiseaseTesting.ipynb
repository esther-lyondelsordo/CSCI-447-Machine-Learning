Part B: plot probability as a function of specificity

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
```

```python
spec = 98
def covidGivenPositive(spec):
    denom = (70*.7) + ((100-spec)*99.3)
    return (70*.7)/denom *100
```

```python
covidGivenPositive(spec)
```

```python
X = np.arange(0,100,1)
Y = covidGivenPositive(X)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_ylabel('P(having COVID|positive test result) * 100')
ax.set_xlabel('Specificity (out of 100)')
plt.plot(X,Y)
plt.axvline(98, 0, 40, color = 'r')
plt.text(98.1,0,'98',rotation=90)
plt.show()
```
